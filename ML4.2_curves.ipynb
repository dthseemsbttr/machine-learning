```python
from sklearn.datasets import load_breast_cancer
```

```python
breast_cancer = load_breast_cancer()
```

```python
X = breast_cancer.data
y = breast_cancer.target
X_2d = X[:, :2]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_2d, y, random_state=0)
```

```python
from sklearn.svm import SVC

clf = SVC(C=0.01) #rbf - default
clf.fit(X_train, y_train)
print(clf.score(X_train, y_train))
print(clf.score(X_test, y_test))
```

```python
import matplotlib.pyplot as plt
import numpy as np

training_scores = []
test_scores = []
numbers = []

for i in np.linspace(20, len(X_train), 20):
    clf = SVC(C=0.01, kernel='rbf')
    X = X_train[:int(i)]
    y = y_train[:int(i)]
    clf.fit(X, y)
    numbers.append(int(i))
    training_scores.append(clf.score(X_train, y_train))
    test_scores.append(clf.score(X_test, y_test))

plt.figure(figsize=(10, 5))
plt.plot(numbers, training_scores, label="training scores")
plt.plot(numbers, test_scores, label="test scores")
plt.legend(loc="best")
```

```python
X = breast_cancer.data
y = breast_cancer.target
X_2d = X[:, :2]
X_train, X_test, y_train, y_test = train_test_split(X_2d, y, random_state=0)
```

```python
from sklearn.svm import SVC

clf = SVC(C=1) #rbf - default
clf.fit(X_train, y_train)
print(clf.score(X_train, y_train))
print(clf.score(X_test, y_test))
```

```python
def graph(c):
    training_scores = []
    test_scores = []
    numbers = []

    for i in np.linspace(20, len(X_train), 20):
        clf = SVC(C=c, kernel='rbf')
        X = X_train[:int(i)]
        y = y_train[:int(i)]
        clf.fit(X, y)
        numbers.append(int(i))
        training_scores.append(clf.score(X_train, y_train))
        test_scores.append(clf.score(X_test, y_test))

    plt.figure(figsize=(10, 5))
    plt.plot(numbers, training_scores, label="training scores")
    plt.plot(numbers, test_scores, label="test scores")
    plt.legend(loc="best")
    
    return plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
training_scores = []
test_scores = []
numbers = []

for i in np.linspace(20, len(X_train), 20):
    lg = LogisticRegression()
    X = X_train[:int(i)]
    y = y_train[:int(i)]
    lg.fit(X, y)
    numbers.append(int(i))
    training_scores.append(lg.score(X_train, y_train))
    test_scores.append(lg.score(X_test, y_test))

plt.figure(figsize=(10, 5))
plt.plot(numbers, training_scores, label="training scores")
plt.plot(numbers, test_scores, label="test scores")
plt.legend(loc="best")
```

```python
C = 0.0001

training_scores = []
test_scores = []
numbers = []
C_list = []

for i in range(11):
    C_list.append(C * 10 ** i)
    clf = SVC(C=C * 10 ** i, kernel='rbf')
    X = X_train
    y = y_train
    clf.fit(X, y)
    numbers.append(int(i))
    training_scores.append(clf.score(X_train, y_train))
    test_scores.append(clf.score(X_test, y_test))

plt.figure(figsize=(10, 5))
plt.plot(numbers, training_scores, label="training scores")
plt.plot(numbers, test_scores, label="test scores")
plt.xticks(numbers, labels=C_list)
plt.legend(loc="best")
plt.xlabel('C values')
plt.ylabel('Accuracy')
plt.show()
```

```python
training_scores = []
test_scores = []
numbers = []
maxiter = []

for i in range(1, 7):
    clf = LogisticRegression(max_iter=10**i)
    X = X_train
    y = y_train
    clf.fit(X, y)
    numbers.append(i)
    maxiter.append(10**i)
    training_scores.append(clf.score(X_train, y_train))
    test_scores.append(clf.score(X_test, y_test))

plt.figure(figsize=(10, 5))
plt.plot(numbers, training_scores, label="training scores")
plt.plot(numbers, test_scores, label="test scores")
plt.legend(loc="best")
plt.xticks(numbers, labels=maxiter)
plt.xlabel('Max iter')
plt.ylabel('Accuracy')
plt.show()
```

```python
from sklearn.linear_model import SGDClassifier

training_scores = []
test_scores = []
learning_rate = []

for i in ['constant', 'optimal', 'invscaling', 'adaptive']:
    clf = SGDClassifier(eta0=0.01, learning_rate=i)
    X = X_train
    y = y_train
    clf.fit(X, y)
    learning_rate.append(i)
    training_scores.append(clf.score(X_train, y_train))
    test_scores.append(clf.score(X_test, y_test))

plt.figure(figsize=(10, 5))
plt.plot(training_scores, label="training scores")
plt.plot(test_scores, label="test scores")
plt.legend(loc="best")
plt.xticks(range(len(learning_rate)), labels=learning_rate)
plt.xlabel('Max iter')
plt.ylabel('Accuracy')
plt.show()
```
