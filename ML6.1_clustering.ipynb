```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import silhouette_score, calinski_harabasz_score
import matplotlib.cm as cm
from sklearn.cluster import KMeans
%matplotlib inline
```

# №1

```python
from sklearn.preprocessing import LabelEncoder
df = pd.read_csv('Mall_Customers.csv', index_col=0)
df['Genre'] = LabelEncoder().fit_transform(df['Genre'])
X = df
```

```python
df
```

# №2

```python
df.isna().sum()
```

# №3

```python
sns.scatterplot(x='Annual Income (k$)' , y = 'Spending Score (1-100)',data=df , hue='Genre')
```

# №4

```python
x = np.array(df['Annual Income (k$)']).reshape(-1,1)
OMP_NUM_THREADS=1

wcss = []
for i in range(1,11):
    k_means = KMeans(n_clusters=i,random_state=42)
    k_means.fit(x)
    wcss.append(k_means.inertia_)
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
```

# №5

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

# №6

```python
plt.scatter(x, df['Spending Score (1-100)'], c=y_kmeans, s=20, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 2], centers[:, 3], c='black', s=200, alpha=0.5)
```

```python
kmeans.cluster_centers_
```

# №7

```python
plt.scatter(df['Age'], df['Spending Score (1-100)'], c=y_kmeans, s=20, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 1], centers[:, 3], c='black', s=200, alpha=0.5)
```

# №8

```python
labels = kmeans.labels_
```

```python
from sklearn import metrics
metrics.silhouette_score(X, labels, metric='euclidean')
```

```python
metrics.calinski_harabasz_score(X, labels)
```
