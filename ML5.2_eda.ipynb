# Доп Задания
# №1

```python
Y = df['satisfaction']
X = df.drop('satisfaction', axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, 
                                                    test_size=0.2)
```

```python
%%timeit -n1 -r1 -o
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
cls = LogisticRegression(max_iter = 2000)
cls.fit(x_train, y_train)
y_pred = cls.predict(x_test)
print(cls_accuracy := metrics.accuracy_score(y_test, y_pred))
print(cls_precision := metrics.precision_score(y_test, y_pred))
print(cls_recall := metrics.recall_score(y_test, y_pred))
print(cls_f1 := metrics.f1_score(y_test, y_pred))
```

```python
timecls = _.average
timecls
```

```python
%%timeit -n1 -r1 -o
from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier()
knc.fit(x_train, y_train)
y_pred = knc.predict(x_test)
print(knc_accuracy := metrics.accuracy_score(y_test, y_pred))
print(knc_precision := metrics.precision_score(y_test, y_pred))
print(knc_recall := metrics.recall_score(y_test, y_pred))
print(knc_f1 := metrics.f1_score(y_test, y_pred))
```

```python
timeknc = _.average
timeknc
```

```python
%%timeit -n1 -r1 -o
from sklearn.neural_network import MLPClassifier
mlpc = MLPClassifier(max_iter=1000)
mlpc.fit(x_train, y_train)
y_pred = mlpc.predict(x_test)
print(mlpc_accuracy := metrics.accuracy_score(y_test, y_pred))
print(mlpc_precision := metrics.precision_score(y_test, y_pred))
print(mlpc_recall := metrics.recall_score(y_test, y_pred))
print(mlpc_f1 := metrics.f1_score(y_test, y_pred))
```

```python
timemlpc = _.average
timemlpc
```

```python
%%timeit -n1 -r1 -o
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
y_pred = dtc.predict(x_test)
print(dtc_accuracy := metrics.accuracy_score(y_test, y_pred))
print(dtc_precision := metrics.precision_score(y_test, y_pred))
print(dtc_recall := metrics.recall_score(y_test, y_pred))
print(dtc_f1 := metrics.f1_score(y_test, y_pred))
```

```python
timedtc = _.average
timedtc
```

```python
timecol = [timecls, timeknc, timemlpc, timedtc]
accuracycol = [cls_accuracy, knc_accuracy, mlpc_accuracy, dtc_accuracy]
precisioncol = [cls_precision, knc_precision, mlpc_precision, dtc_precision]
recallcol = [cls_recall, knc_recall, mlpc_recall, dtc_recall]
f1col = [cls_f1, knc_f1, mlpc_f1, dtc_f1]
metrics_columns = ['cls', 'knc', 'mlpc', 'dtc']
timecol
```

```python
df_metrics = pd.DataFrame([timecol, accuracycol, precisioncol, recallcol, f1col], index = ['time', 'accuracy', 'precision', 'recall', 'f1'])
df_metrics.columns = metrics_columns
```
