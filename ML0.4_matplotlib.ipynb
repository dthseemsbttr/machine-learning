```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('train_1.csv')
```

```python
df
```

```python
rooms_count = df['Rooms'].value_counts()
plt.pie(rooms_count, labels=rooms_count.index, explode=[0.2]+[0]*(len(rooms_count)-1))
```

```python
plt.hist(df['Price'])
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
ax[0,0].scatter(x = df['Price'], y = df['Rooms'])
ax[0,1].scatter(x = df['Price'], y = df['Square'])
ax[1,0].scatter(x = df['Price'], y = df['HouseFloor'])
ax[1,1].scatter(x = df['Price'], y = df['HouseYear'])
```

```python
sns.kdeplot(df['Price'])
```

```python
sns.kdeplot(df['Price'], df['HouseFloor'])
plt.ylim(0, 40)
```

```python
plt.boxplot(df['Square'])
```

```python
sns.pairplot(df[['Rooms', 'Square', 'HouseFloor', 'HouseYear', 'Price']]).map_lower(sns.kdeplot, levels=20)
```

```python
sns.heatmap(df[['Rooms', 'Square', 'HouseFloor', 'HouseYear', 'Price']].corr())
```

```python
sns.jointplot(df['Square'], df['Price'])
```
